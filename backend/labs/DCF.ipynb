Discounted Cash Flow
-
$$
DCF = \sum_{t=1}^{n}\frac{FCF_t}{(1+r)^t} + \frac{TV}{(1+r)^n}
$$

$FCF_t$ = Free Cash Flow in period $t$  
$TV$ = Terminal Value  
${r}$ = The discount rate (WACC,Weighted Average Cost of Capital) 

$$
TV = \frac{FCF_n\times(1+g)}{r-g}
$$

$FCF_n$ = Free Cash Flow in the final year of the projection period.  
$g$ = Perpetual growth rate of the cash flows (typically matched to long-term GDP growth or inflation).
